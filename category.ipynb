```python
final_file = os.path.join(folder_path, file_name)
print("final_file = ", final_file)

start_time = time.time()
master_df = pd.read_excel(final_file, sheet_name="PMS_Master_Data_1")
print(master_df.head())

print("master_df = ", master_df.shape)
print("Time Taken = ", time.time() - start_time)
```

```python
master_df = master_df[master_df['Scope'] == 'In Scope']
print("master_df = ", master_df.shape)
```

```python
cursor = connection.cursor()

# Define the insert statement
insert_query = """
INSERT INTO category_master (
    NSSID, domain_dict, circle, site_type_vi, site_type_infra, frequency, 
    proposed_frequency, synergy, priority, site_name, vendor_name, 
    lat_long, domain, node_name, sub_category, sub_category1
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""



unique_circle = master_df['CIRCLE_CODE'].unique().tolist()
print("unique_circle = ", unique_circle)

for circle in unique_circle:
    filtered_circle = master_df[master_df['CIRCLE_CODE']==circle]
    print("filtered_circle = ", circle, filtered_circle.shape)

    # circle = "GUJ"
    filtered_circle = master_df[master_df['CIRCLE_CODE']==circle]
    print("filtered_circle = ", circle, filtered_circle.shape)


    nssid_count_df = filtered_circle.groupby(['NSSID', 'DOMAIN']).size().reset_index(name='count')
    nssid_count_domain = nssid_count_df['DOMAIN'].unique().tolist()
    # print("nssid_count_domain: ", nssid_count_domain)
    
    # sitetype_to_find = ["RAN-BTS", "MICROWAVE", "OPTICS", "IPMPLS-Core", "RAN-BSC", "RNC", "DWDM"]
    # print("sitetype_to_find = ", sitetype_to_find)
    for item in site_ownership:
        print()
        # print(item['site_type'])
        sitetype_to_find = item['site_type']
        site_type_vi = item['site_type_vi']
        site_type_infra = item['site_type_infra']
        frequency = item['frequency']
        proposed_frequency = item['proposed_frequency']
        synergy = item['synergy']
        priority = item['priority']
        
        common_domains = list(set(nssid_count_domain) & set(sitetype_to_find))
        # print("Common domains in both lists: ", common_domains)
        
        
        filtered_nss_df = nssid_count_df[nssid_count_df['DOMAIN'].isin(common_domains)]
        domain_filter_df = filtered_nss_df.groupby('NSSID').filter(lambda x: set(x['DOMAIN']) == set(common_domains))
        
        final_ndd_count_df = domain_filter_df.pivot(index='NSSID', columns='DOMAIN', values='count').reset_index()
        # print("final_ndd_count_df = ", final_ndd_count_df)
    
        for index, row in final_ndd_count_df.iterrows():
            NSSID = row['NSSID']
            domain_dict = row.drop('NSSID').to_dict()
    
        
            matched_master_rows = master_df[
                (master_df['CIRCLE_CODE'] == circle) &
                (master_df['NSSID'] == NSSID)
            ]
    
            site_name = matched_master_rows['SITE_NAME'].iloc[0]
            vendor_name = matched_master_rows['VENDOR_NAME'].iloc[0]
            lat_long = matched_master_rows['LATLONG'].iloc[0]
            domain = matched_master_rows['DOMAIN'].iloc[0]
            node_name = matched_master_rows['NODE_NAME'].iloc[0]
            sub_category = matched_master_rows['Sub Category'].iloc[0]
            sub_category1 = matched_master_rows['Sub Category 1'].iloc[0]
    
            domain_dict = str(domain_dict)
            synergy = str(synergy)
    
            # print("NSSID = ", type(NSSID))
            # print("domain_dict = ", type(domain_dict))
            # print("circle = ", type(circle))
            # print("site_type_vi = ", type(site_type_vi))
            # print("site_type_infra = ", type(site_type_infra))
            # print("frequency = ", type(frequency))
            # print("proposed_frequency = ", type(proposed_frequency))
            # print("synergy = ", type(synergy))
            # print("priority = ", type(priority))
            # print("site_name = ", type(site_name), site_name)
            # print("vendor_name = ", type(vendor_name), vendor_name)
            # print("lat_long = ", type(lat_long), lat_long)
            # print("domain = ", type(domain), domain)
            # print("node_name = ", type(node_name), node_name)
            # print("sub_category = ", type(sub_category), sub_category)
            # print("sub_category1 = ", type(sub_category1), sub_category1)
            
        
            # Data to be inserted
            data_to_insert = (
                NSSID, domain_dict, circle, site_type_vi, site_type_infra, frequency, 
                proposed_frequency, synergy, priority, site_name, vendor_name, 
                lat_long, domain, node_name, sub_category, sub_category1
            )
        
            # Execute the insert statement
            cursor.execute(insert_query, data_to_insert)
    
            # Commit the transaction
            connection.commit()

# Close the cursor and connection
cursor.close()
connection.close()

print("Data inserted successfully into category_master table")




# # Ensure the DOMAIN column is a single dimension
# if isinstance(filtered_circle.columns, pd.MultiIndex):
#     filtered_circle = filtered_circle.reset_index()

# # Confirm DOMAIN is a simple column
# filtered_circle = filtered_circle[['NSSID', 'DOMAIN']]
# print(filtered_circle)


# # Create the pivot table
# pivot_df = filtered_circle.pivot_table(
#     index='NSSID',       # Rows
#     columns='DOMAIN',    # Columns
#     values='NSSID',      # Values (count of NSSID for each DOMAIN)
#     aggfunc='count',     # Aggregation function
#     fill_value=0         # Replace NaN with 0
# )
# print(pivot_df)

# Save the pivot table to an Excel file
# final_file1 = os.path.join(folder_path, "filtered_pivoted_data.xlsx")
# final_ndd_count_df.to_excel(final_file1, sheet_name='FilteredPivot')
print("Done")

```

```python
import psycopg2

connection = psycopg2.connect(
    host="localhost",
    database="pms_test",
    user="postgres",
    password="admin"
)

cursor = connection.cursor()

insert_query = """
INSERT INTO test1 (column1, column2, column3)
VALUES (%s, %s, %s)
"""

data_to_insert = ('value1', 1, 1)

try:
    cursor.execute(insert_query, data_to_insert)
    connection.commit()
    print("Data inserted successfully")

except Exception as e:
    connection.rollback()
    print(f"Error inserting data: {e}")

finally:
    cursor.close()
    connection.close()
```
